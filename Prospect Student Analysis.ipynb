```r
# add spacing, melt, sort
total <- subset(Funnel, Percentage==100)$Count.Applicant
Funnel$padding <- (total - Funnel$Count.Applicant) / 2
molten <- melt(Funnel[, -3], id.var='Stage')
molten <- molten[order(molten$variable, decreasing = T), ]
molten$Stage <- factor(molten$Stage, levels = rev(Funnel$Stage))

ggplot(molten, aes(x=Stage)) +
  geom_bar(aes(y = value, fill = variable),
           stat='identity', position='stack') +
  geom_text(data=Funnel, 
            aes(y=total/2, label= paste(round(Percentage), '%')),
            color='Black') +
  scale_fill_manual(values = c('Tomato', NA) ) +
  coord_flip() +
  theme(legend.position = 'none') +
  labs(x='stage', y='Percentage')+
ggtitle("Conversion of applicant on each step of the process:")
```
